````python
products = ["tshirt", "mug", "hat", "book", "keychain"]

def initialize_inventory(products):
    inventory = {product: int(input(f"Enter the quantity of {product}s available: ")) for product in products}
    return inventory

def get_customer_orders():
    num_orders = int(input("Enter the number of customer orders: "))
    orders = [input("Enter the name of a product that a customer wants to order: ").lower() for _ in range(num_orders)]
    return orders

def update_inventory(inventory, customer_orders):
    for order in customer_orders:
        if order in inventory and inventory[order] > 0:
            inventory[order] -= 1
    inventory = {k: v for k, v in inventory.items() if v > 0}
    return inventory

def calculate_total_price(customer_orders):
    prices = {product: float(input(f"Enter the price of {product}: ")) for product in customer_orders}
    total = sum(prices.values())
    return total

inventory = initialize_inventory(products)
customer_orders = get_customer_orders()

total_orders = len(customer_orders)
unique_orders = len(set(customer_orders))
percentage_unique = (unique_orders / len(products)) * 100

print("\nOrder Statistics:")
print("Total Products Ordered:", total_orders)
print("Percentage of Unique Products Ordered:", round(percentage_unique, 2))

inventory = update_inventory(inventory, customer_orders)
print("\nUpdated Inventory:")
for item, qty in inventory.items():
    print(item, ":", qty)

total_price = calculate_total_price(customer_orders)
print("Total Price:", total_price)
````
